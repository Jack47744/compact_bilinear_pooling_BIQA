###LIVE Database
H. R. Sheikh, M. F. Sabir, and A. C. Bovik, “A statistical evaluation
of recent full reference image quality assessment algorithms,” IEEE
Transactions on Image Processing, vol. 15, no. 11, pp. 3440–3451, Nov.
2006.

```python
!unzip '/content/drive/My Drive/image_quality_dataset/LIVE_RAW/databaserelease2.zip'
```

```python
live_meta = pd.read_excel('/content/drive/My Drive/image_quality_dataset/LIVE_RAW/LIVE_SCORE.xlsx')
distortion_type = ['jp2k', 'jpeg', 'wn', 'gblur', 'fastfading']
start_idx = 0
image_score_map = {}
image_number_each_distortion = [227, 233, 174, 174, 174]
for idx, dis in enumerate(distortion_type):
    for i in range(image_number_each_distortion[idx]):
        score = live_meta.iloc[start_idx, 1]
        start_idx += 1
        image_new_name = 'LIVE_'+dis+'_'+str(i+1)+'.bmp'
        img_name = 'img'+str(i+1)+'.bmp'
        old_img_path = '/content/databaserelease2/'+dis+'/'
        new_img_path = '/content/drive/My Drive/image_quality_dataset/LIVE/Train/'
        if score > 0:
            print(score)
            !mv {old_img_path+img_name} {old_img_path+image_new_name}
            !cp {old_img_path+image_new_name} '/content/drive/My Drive/image_quality_dataset/LIVE/Train/'
            image_score_map[image_new_name] = score
```

Select 20% to be a test set.

```python
path = '/content/drive/My Drive/image_quality_dataset/LIVE/Train/'
files = os.listdir(path)
test_img = random.sample(files, int(len(files)*0.2))
for move_img in test_img:
    !mv "/content/drive/My Drive/image_quality_dataset/LIVE/Train/"{move_img} '/content/drive/My Drive/image_quality_dataset/LIVE/Test/'
```

Export csv file

```python
img_name = [k for k in image_score_map.keys()]
img_score = [k for k in image_score_map.values()]
d = {'img_name':img_name, 'img_score':img_score}
meta = pd.DataFrame(data=d)
meta.to_csv('/content/drive/My Drive/image_quality_dataset/LIVE/meta.csv', index=False)
```
